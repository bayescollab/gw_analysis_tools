```python
a2 = -0.0520685
a3 = -13.9349
a4 = 0.262694
a5 = 0.993001




def RDphasea4(frequency, mass_1, mass_2, chi1, chi2, f_ring, f_damp):
    
    import bilby
    from astropy import constants as const


    ### const defination
    G_const = const.G.value              # m^3/(kg s^2)
    c_const = const.c.value              # m/s   ### c=1
    M_solar_const = const.M_sun.value    # kg
    
    
    
    ### Constructing the EFT correction
    freq = np.array(frequency)
    M_tot = mass_1 + mass_2
    M_tot_kg = M_tot*M_solar_const

    eta = mass_1*mass_2/(M_tot*M_tot)
    
    a0 = -15.6903
    a1 = 82.5651
    
    print(a2)
    print(a3)
    print(a4)
    print(a5)
    print(eta)
    # a2 = -0.0520685
    # a3 = -13.9349
    # a4 = 0.262694
    # a5 = 0.993001

    Mf = (M_tot_kg*freq*G_const)/np.power(c_const,3) # dimensionless
    #print(Mf)

    
    return 1/eta * (a4*np.arctan( (frequency-a5*f_ring) / f_damp))


def RDphasea3(frequency, mass_1, mass_2, chi1, chi2, f_ring, f_damp):
    
    import bilby
    from astropy import constants as const


    ### const defination
    G_const = const.G.value              # m^3/(kg s^2)
    c_const = const.c.value              # m/s   ### c=1
    M_solar_const = const.M_sun.value    # kg
    
    
    
    ### Constructing the EFT correction
    freq = np.array(frequency)
    M_tot = mass_1 + mass_2
    M_tot_kg = M_tot*M_solar_const

    eta = mass_1*mass_2/(M_tot*M_tot)
    
    a0 = -15.6903
    a1 = 82.5651
    
    print(a2)
    print(a3)
    print(a4)
    print(a5)
    print(eta)
    # a2 = -0.0520685
    # a3 = -13.9349
    # a4 = 0.262694
    # a5 = 0.993001

    Mf = (M_tot_kg*freq*G_const)/np.power(c_const,3) # dimensionless
    #print(Mf)

    
    return 1/eta * (4/3*a3*np.power(Mf,3/4))



phase_a4 = RDphasea4(phase_freq, 35, 25, 0, 0, 297.23, 22.9785)
phase_a3 = RDphasea3(phase_freq, 35, 25, 0, 0, 297.23, 22.9785)

pylab.figure()
pylab.plot(phase_freq, phase_a4)
#pylab.plot(phase_freq, phase_EFT_RD, linestyle="--")
pylab.axvline(x=f2, linestyle="--")
#pylab.xlim(f2-10,f2+10)
#pylab.ylim(-60,-45)
pylab.xlabel('Frequency (Hz)')
pylab.ylabel('phase')
#pylab.legend(["phase_EFT_RD", "phase_RD"])
pylab.show()


pylab.figure()
pylab.plot(phase_freq, phase_a3)
#pylab.plot(phase_freq, phase_EFT_RD, linestyle="--")
pylab.axvline(x=f2, linestyle="--")
#pylab.xlim(f2-10,f2+10)
#pylab.ylim(-60,-45)
pylab.xlabel('Frequency (Hz)')
pylab.ylabel('phase')
#pylab.legend(["phase_EFT_RD", "phase_RD"])
pylab.show()

    
```
