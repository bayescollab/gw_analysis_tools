# Waveform Visualization



```python
import numpy as np
import matplotlib.pyplot as plt
import corner
from scipy.stats import multivariate_normal
```

## Load in data

```python
strains = np.loadtxt("../data/responses.csv",delimiter=',')
polarizations = np.loadtxt("../data/polarizations.csv",delimiter=',')
phases = np.loadtxt("../data/phases.csv",delimiter=',')
HStrain = strains[1,:] + 1j*strains[2,:]
LStrain = strains[3,:] + 1j*strains[4,:]
VStrain = strains[5,:] + 1j*strains[6,:]
freqStrain = strains[0,:]

phaseplus = phases[1,:] 
phasecross = phases[2,:] 
freqPhase = phases[0,:]
hplus = polarizations[1,:] + 1j* polarizations[2,:] 
hcross = polarizations[3,:] + 1j* polarizations[4,:] 
freqPol = polarizations[0,:]
print(phaseplus-phasecross)
```

```python
plt.plot(freqPhase, phaseplus)
plt.plot(freqPhase, phasecross,linestyle=':')
plt.xscale('log')
plt.show()
plt.close()
```

```python
plt.loglog(freqStrain, np.abs(HStrain))
plt.loglog(freqStrain, np.abs(LStrain))
plt.loglog(freqStrain, np.abs(VStrain))
# plt.loglog(freqPhase, phasecross,linestyle=':')
plt.show()
plt.close()
```

```python
plt.loglog(freqPol, np.abs(hplus))
plt.loglog(freqPol, np.abs(hcross))
# plt.loglog(freqPhase, phasecross,linestyle=':')
plt.show()
plt.close()
```
