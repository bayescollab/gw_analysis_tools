```python
import matplotlib.pyplot as plt
import csv
import numpy as np
import math
```

```python
rows = []
with open("sys_err_data.csv") as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        rows.append(row)
```

```python
snr_arr = []
sys_err_arr = []

for i in rows:
    snr_arr.append(eval(i[0]))
for i in range(11):
    temp = []
    for j in rows:
        temp.append(eval(j[i+1]))
    sys_err_arr.append(temp)
```

```python
snr = np.array(snr_arr)
sys_err = []
for i in range(11):
    temp = np.array(sys_err_arr[i])
    sys_err.append(temp)
#sys_err_1 = np.array(sys_err_arr[0])
```

```python
param_info = ["RA", "DEC", "psi", "phiRef", "tc", "iota_L", "ln DL", "ln chirpmass", "eta", "chi1", "chi2"]

for i in range(11):
    plt.plot(snr, sys_err[i]/sys_err[i][19], label = param_info[i])
#fig, axs = plt.subplots(4,3)
#for i in range(11):
#    axs[i%4, i%3].plot(snr, sys_err[i]/sys_err[i][19])
plt.legend()
plt.show()
```
