```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import spearmanr
```

```python
data1 = np.sin(np.linspace(1,100,1000))
data2 = np.cos(np.linspace(1,100,1000))
```

```python
data1 = np.random.rand(1000)
data2=np.random.rand(1000)
```

```python
print(np.correlate(data1,data2))
print(spearmanr(data1,data2))
corr = np.correlate(data1,data2,"full")

plt.plot(corr)
plt.show()
plt.close()
plt.plot(data1)
plt.plot(data2)
plt.show()
plt.close()
```
